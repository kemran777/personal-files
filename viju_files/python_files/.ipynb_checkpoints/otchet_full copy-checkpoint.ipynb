```python
# DAU
def daily(table='2023-03-24'):
    dau = pd.read_sql(f'''
    select 
        count(distinct profile_id) as count
    from datamarts.dash_table_distr
    where client_type not in ('backend', 'testing')
    and date = '{table}'
    ''', con=al)
    return dau['count'].values[0]


def registrations(date='2023-01-10'):
    regs = pd.read_sql(f'''
    select 
        count(distinct id) as count
    from users u
    where
        user_type = 'regular'
        and (
            u.email not ilike '%%@test%%' and u.email not ilike '%%@viasat%%' or
            (u.email is null and u.phone_number is not null)
        )
        and cast(u.created_at as date) = '{date}'
    ''', con=con)
    return regs['count'].values[0]


def new_trials(date='2023-01-10'):
    new = pd.read_sql(f'''
    select 
        count(DISTINCT user_id) as count
    from datamarts.marketing_dash_distr 
    where created_at::date = '{date}'
    and promo_type = 'no_promo'
    and state in ('normal_period','trial','canceled')
    ''',con=al)
    return new['count'].values[0]


def new_codes(date='2022-09-25'):
    new = pd.read_sql(f'''
    select 
        count(DISTINCT user_id) as count
    from datamarts.marketing_dash_distr 
    where created_at::date = '{date}'
    and promo_type = 'promo'
    and state in ('normal_period','trial','canceled')
    ''',con=al)
    return new['count'].values[0]


def churned_trials(date='2023-01-02'):
    churned = pd.read_sql(f'''
    select 
        count(DISTINCT user_id) as count
    from datamarts.marketing_dash_distr 
    where ends_at::date = '{date}'
    and promo_type = 'no_promo'
    and state in ('canceled')
    and payer = 0
    ''',con=al)
    return churned['count'].values[0]


def prolonged_trials(date='2023-01-02'):
    new = pd.read_sql(f'''
    select 
        count(DISTINCT user_id) as count
    from datamarts.marketing_dash_distr 
    where first_prolong_date::date = '{date}'
    and promo_type = 'no_promo'
    and state in ('normal_period','trial','canceled')
    and payer = 1
    ''',con=al)
    return new['count'].values[0]

def prolongs_codes(date):
    new = pd.read_sql(f'''
    select 
        count(DISTINCT user_id) as count
    from datamarts.marketing_dash_distr 
    where first_prolong_date::date = '{date}'
    and promo_type = 'promo'
    and state in ('normal_period','trial','canceled')
    and payer = 1
    ''',con=al)
    return new['count'].values[0]


def new_codes_b2b(date='2022-09-25'):
    new = pd.read_sql(f'''
    select 
        count(DISTINCT user_id) as count
    from datamarts.marketing_dash_distr 
    where bonus_activated_at::date = '{date}'
    and promo_type = 'cards'
    and state in ('normal_period','trial','canceled')
    ''',con=al)
    return new['count'].values[0]


def active_codes_recurrent(date='2023-01-02'):
    new = pd.read_sql(f'''
    select 
        count(DISTINCT user_id) as count
    from datamarts.marketing_dash_distr 
    where 
    promo_type = 'promo'
    and state in ('normal_period','trial','canceled')
    and payer = 0
    and renewal_off_date > '{date}'
    ''',con=al)
    return new['count'].values[0]


def churned_codes(date='2023-01-02'):
    new = pd.read_sql(f'''
    select 
        count(DISTINCT user_id) as count
    from datamarts.marketing_dash_distr 
    where ends_at::date = '{date}'
    and promo_type = 'promo'
    and state in ('normal_period','trial','canceled')
    and payer = 0
    ''',con=al)
    return new['count'].values[0]


def dau_paid(date):
    table = 'events_'+date.replace('-','_')
    paid = pd.read_sql(f'''
    select 
        distinct profile_id
    from datamarts.marketing_dash_distr 
    where payer = 1
    and promo_type != 'cards'
    and state in ('normal_period','trial','canceled', 'grace_period')
    and ends_at >= '{date}'
    ''',con=al)
    temp = pd.read_sql(f'''
        select 
            distinct profile_id
        from {table}
        where 
            client_type != 'backend'
    ''', con=engine)
    return temp[temp['profile_id'].isin(paid['profile_id'].astype(str).unique())].shape[0]


def dau_b2b(date):
    table = 'events_'+date.replace('-','_')
    subs = pd.read_sql(f'''
    select 
        distinct profile_id
    from datamarts.marketing_dash_distr 
    where 
    promo_type = 'cards'
    and state in ('normal_period','trial','canceled')
    and bonus_ends_at >= '{date}'
    ''',con=al)
    temp = pd.read_sql(f'''
        select 
            distinct profile_id
        from {table}
        where
        client_type !='backend'
    ''', con=engine)
    return temp[temp['profile_id'].astype(str).isin(subs['profile_id'].astype(str).unique())].shape[0]

def daily_vis(table='2023-03-24'):
    table = 'events_'+table.replace('-','_')
    testers = pd.read_sql('''
    select 
        id, visitor_id
    from users
    where deleted_at is null
    and user_type = 'tester'
    ''', con=con)
    testers['id'] = testers['id'].astype(str)
    dau = pd.read_sql(f'''
    select 
        distinct visitor_id
    from {table}
    where client_type != 'backend'
    ''', con=engine)
    return dau[~dau['visitor_id'].isin(testers['visitor_id'].unique())]['visitor_id'].nunique()


def renewal_off(date):
    df = pd.read_sql(f'''
    select count(distinct user_id) as count
    from datamarts.marketing_dash_distr mdd 
    WHERE 
    renewal_off_date::date = '{date}'
     ''',con=al)
    return df['count'].values[0]


def renewal_off_subs(date):
    df = pd.read_sql(f'''
    select 
	count(distinct user_id) as count
    from datamarts.marketing_dash_distr mdd 
    WHERE 
    renewal_off_date::date = '{date}'
    and payer = 0
    and len < 1
    and state in ('trial','normal_period','canceled','grace_period')
     ''',con=al)
    return df['count'].values[0]


def mau(date):
    df = pd.read_sql(f'''
    select count(DISTINCT profile_id) as count
    from superset.events_distr
    where   
    utc_timestamp::date between date_sub('{date}'::date, interval 29 day) and '{date}'
    and client_type != 'backend' ''',con=al)
    return df['count'].values[0]


def wau(date):
    df = pd.read_sql(f'''
    select count(DISTINCT profile_id) as count
    from superset.events_distr
    where 
    utc_timestamp::date between date_sub('{date}'::date, interval 6 day) and '{date}'
    and client_type != 'backend' ''',con=al)
    return df['count'].values[0]
```

##### B2B subs

```python
subs = pd.read_sql('''
select
    u.id,
        s.id as sub,
        u.email,
        u.phone_number,
        cast(u.created_at as date) as reg_date,
        s.state,
        recurrent,
        s.offer_id,
        cast(s.created_at as date),
        cast(s.ends_at as date),
        cast(s.ends_at as date) - cast(s.created_at as date) as dur,
        pc.title as promo_code,
        bp.title as bonus_title,
        ub.type as bonus_type,
        cast(ub.activated_at as date),
        cast(ub.starts_at as date),
        cast(ub.ends_at as date) as bonus_ends,
        cast(ub.ends_at as date) - cast(ub.starts_at as date) as bonus_dur,
        r.pays[2:] as all_prolong_dates,
        case when len > 0
            then
                to_json(r.pays)->-1 
            else 
                NULL
        end as last_prolong_date,
        (
            SELECT 
                string_agg(to_char(invoices.paid_at AT TIME ZONE 'UTC' AT TIME ZONE 'Europe/Moscow', 'YYYY-MM-DD'), ', ')
            FROM 
                (SELECT * FROM invoices ORDER BY subscription_id, paid_at) invoices
            WHERE 
                invoices.deleted_at IS NULL
                AND invoices.paid_at IS NOT NULL
                AND invoices.payment_target = 'trial_period'
                AND invoices.subscription_id = s.id
            GROUP BY invoices.subscription_id
        )            
        AS trial_date,
        r.pays,
        r.money,
        r.durations,
        targets,
        r.len
FROM subscriptions s
INNER JOIN users u on (u.id = s.user_id AND u.deleted_at IS NULL AND u.user_type = 'regular' and u.vipplay=False)
left join user_bonuses ub on (ub.user_id = u.id and ub.deleted_at is null and ub.initiator_type = 'PromoCodeActivation' and ub.activated_at is not null)
left join promo_code_activations pca on pca.user_bonus_id = ub.id and pca.deleted_at is null
left join promo_codes pc on pc.id = pca.promo_code_id and pc.deleted_at is null
left join bonus_programs bp on bp.id = pc.bonus_program_id and bp.deleted_at is null
left join 
    (
        select 
            user_id,
            array_remove(array_agg(cast(i2.paid_at as date) order by i2.paid_at),NULL) as pays,
            array_remove(array_agg(i2.payment_target order by i2.paid_at),NULL) as targets,
            array_remove(array_agg(i2.price_cents order by i2.paid_at),NULL) as money, -- список всех сумм из счетов,
            array_remove(array_agg(cast(i2.period_end as date) - cast(i2.paid_at as date) order by i2.paid_at),NULL) as durations,
            cardinality(array_remove(array_agg(cast(i2.paid_at as date)),NULL)) - 1 as len -- сколько было реальных оплат
        from subscriptions s2 
        left join invoices i2 on (i2.subscription_id = s2.id and i2.deleted_at is null and i2.paid_at is not null)
        where s2.deleted_at is null
        group by user_id
    ) r on r.user_id = u.id
where 
    s.deleted_at is null
    and state != 'initial'
    and pca.multiple = False
    and ub.type = 'PromoCodes::Bonus'
    and (u.email not like '%%@test%%' or phone_number is not null)
    and cast(u.created_at as date) >= '2022-01-01'
''',con=con)
print(subs.shape)
subs.head(3)
```

```python
act = pd.DataFrame()
act['dt'] = pd.date_range(start='08/12/2024', end='08/19/2024')
act['dt'] = act['dt'].apply(lambda x: x.to_pydatetime())
act['active_codes'] = 0
act['cards1m'] = 0
act['cards45d'] = 0
act['cards3m'] = 0
act['cards6m'] = 0
act['cards12m'] = 0
act['cards24m'] = 0
res['starts_at'] = pd.to_datetime(res['starts_at'])
res['bonus_ends'] = pd.to_datetime(res['bonus_ends'])

for i, row in tqdm_notebook(act.iterrows()):
    for j, r in res.iterrows():
        if (row['dt'] >= (r['starts_at']))  & ((r['bonus_ends']) >= row['dt']):
            act.loc[i, 'active_codes'] += 1
            if res['bonus_dur'][j] >=27 and res['bonus_dur'][j]<=35:
                act.loc[i, 'cards1m'] += 1
            if (res['bonus_dur'][j] >=88 and res['bonus_dur'][j]<=93):
                act.loc[i, 'cards3m'] += 1
            if (res['bonus_dur'][j] >=178 and res['bonus_dur'][j]<=185):
                act.loc[i, 'cards6m'] += 1
            if (res['bonus_dur'][j] >=363 and res['bonus_dur'][j]<=367):
                act.loc[i, 'cards12m'] += 1
            if (res['bonus_dur'][j] ==45):
                act.loc[i, 'cards45d'] += 1
            if (res['bonus_dur'][j] >=380):
                act.loc[i, 'cards24m'] += 1
act.T
```
