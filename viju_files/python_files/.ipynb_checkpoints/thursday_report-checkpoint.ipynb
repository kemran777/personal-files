## B2C graph

```python
df1 = pd.read_sql('''
--
                 select
                    *
                 from datamarts.marketing_dash_distr
                 where 
                 payer = 1
 ''',con=al)

act = pd.DataFrame()
act['dt'] = pd.date_range(start='04/01/2024', end='08/01/2024')

act['subs'] = 0

df1['first_prolong_date'] = pd.to_datetime(df1['first_prolong_date'])
df1['ends_at'] = pd.to_datetime(df1['ends_at'])

df1 = df1[df1['first_prolong_date'].notna() & df1['ends_at'].notna()]

import numpy as np

dates = act['dt'].values
start_dates = df1['first_prolong_date'].values
end_dates = df1['ends_at'].values

subs = np.zeros_like(dates, dtype=int)
for start, end in zip(start_dates, end_dates):
    subs += (dates >= start) & (dates <= end)

act['subs'] = subs

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
plt.rcParams["figure.dpi"] = 170
# Создаем DataFrame
data = {
    'date': act['dt'],
    'count': act['subs']
}
df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])  # Конвертируем строку в дату

# Построение графика
plt.figure(figsize=(10, 5))
plt.plot(df['date'], df['count'],  linestyle='-', color='blue')

# Установка границ оси x
plt.gca().set_xlim([df['date'].min() - pd.Timedelta(days=15), df['date'].max() + pd.Timedelta(days=15)])

# Форматирование дат на оси X
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

# Улучшение отображения
plt.title('B2C Подписчики')
plt.xlabel('Дата')
plt.ylabel('Количество подписчиков')
plt.legend()
plt.show()

```

```python

df1 = pd.read_sql('''
                 select 
                    date,
                    client_type,
                    count(distinct profile_id) as count
                 from datamarts.dash_table_distr
                 where 
                 date between '2024-06-01' and '2024-08-03'
                 and client_type!='backend'
                 group by 1,2
                 order by 1,2
                 
 ''',con=al)
df1['date']=df1.date.astype('datetime64[ns]')
```

```python

df_dau = pd.read_sql('''
                 select 
                    date,
                    count(distinct profile_id) as count
                 from datamarts.dash_table_distr
                 where 
                 date between '2024-06-01' and '2024-08-03'
                 and client_type!='backend'
                 group by 1
                 order by 1
                 
 ''',con=al)
df_dau['date']=df_dau.date.astype('datetime64[ns]')
```

```python
df_mau = pd.read_sql('''
                 select 
                    count(distinct profile_id) as count
                 from datamarts.dash_table_distr
                 where 
                 date between '2024-06-01' and '2024-06-31'
                 order by 1
                 
 ''',con=al)
```

```python

# Параметры для графика 
plt.figure(figsize=(15, 8))  

# Построение графика для каждого client_type  
for client_type, group in df1.groupby('client_type'):  
    plt.plot(group['date'], group['count'], linestyle='-', marker='o', label=client_type)  

# Форматирование дат на оси X  
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))  

# Улучшение отображения  
sns.set_style("whitegrid")  
plt.title('Количество пользователей по типу клиента')  
plt.xlabel('Дата')  
plt.ylabel('Количество пользователей')  
plt.legend(title='Тип клиента')  
plt.xticks(rotation=45)  # Повернуть метки на оси X для лучшей читаемости  
plt.tight_layout()  # Автоматически подгоняет размеры  
plt.show() 
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
plt.rcParams["figure.dpi"] = 180
# Создаем DataFrame
data = {
    'date': df_dau['date'],
    'count': df_dau['count'],
    
}

df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])

# Define date ranges and colors
date_ranges = {
    '2024-06-01': '2024-07-01',
    '2024-07-01': '2024-08-01',
    '2024-08-01': '2024-08-31'
}
colors = ['#30D5C8', '#0000FF', '#800020']


plt.figure(figsize=(12, 5))
#plt.plot(df['date'], df['count'],  linestyle='-', color='#30D5C8')

for (start_date, end_date), color in zip(date_ranges.items(), colors):
        mask = (df['date'] >= start_date) & (df['date'] <= end_date)
        plt.plot(df.loc[mask, 'date'], df.loc[mask, 'count'], linestyle='-', color=color)

# Formatting
plt.gca().set_xlim([df['date'].min() - pd.Timedelta(days=1), df['date'].max() + pd.Timedelta(days=1)])
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
sns.set_style("whitegrid")
plt.title('DAU')
plt.xlabel('Дата')
plt.ylabel('Количество пользователей')
plt.show()
```
